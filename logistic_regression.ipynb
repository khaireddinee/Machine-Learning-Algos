```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression

#x,y = make_regression(n_samples= 4 , n_features=1, noise=10)
#desired_labels = [0, 0, 1, 1]
#y = np.array(desired_labels).reshape(-1, 1)
x = np.array([[2.5], [3.0], [3.2], [3.5], [4.0], [4.2], [4.5], [5.0], [5.5], [6.0]])
y = np.array([0, 0, 0, 0, 1, 0, 1, 1, 1, 1])
X=x
# sigmoid function
def sigmoid(x):
  return 1/(1+np.exp(-x))

# initialize weights
theta = np.random.randn(2,1)

X= np.hstack((x,np.ones(x.shape)))

# define the model of the algo   //  X.dot(theta) = wx + b = z

def model(X,theta):
  return sigmoid(X.dot(theta))

# cost function that returns the error

def cost_function(X,y,theta):
  m= len(y)
  r= model(X,theta)
  return -(1/m) * np.sum(y * np.log(r) + (1-y) * np.log(1-r) )

# gradient

def grad(X,theta):
  m = len(y)
  r = model(X,theta)
  return (1/m)*X.T.dot(r-y)

# gradient desc

def grad_desc(X,y,theta,lr,n_it):
  cost_history = np.zeros(n_it)
  for i in range(n_it):
    theta = theta - lr * grad(X,theta)
    cost_history[i]=cost_function(X,y,theta)
  return theta,cost_history

theta_final,cost_hist = grad_desc(X,y,theta,0.01,1000)
predictions = model(X,theta_final)
#plt.scatter(x,y)
#theta
#plt.plot(x,predictions,c='r')
```
